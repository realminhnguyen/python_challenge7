## Data Analysis Question 1

The two most important customers of the firm may have been hacked. Verify if there are any fraudulent transactions in their history. For privacy reasons, you only know that their cardholder IDs are 2 and 18.

* Using hvPlot, create a line plot representing the time series of transactions over the course of the year for each cardholder separately. 

* Next, to better compare their patterns, create a single line plot that containins both card holders' trend data.  

* What difference do you observe between the consumption patterns? Does the difference suggest a fraudulent transaction? Explain your rationale in the markdown report.

```python
# loading data for card holder 2 and 18 from the database
# Write the query
query = """
        SELECT
    t.id AS transaction_id,
    t.date,
    t.amount,
    m.name AS merchant_name,
    ch.id AS cardholder_id 
FROM
    transaction t
JOIN
    credit_card cc ON t.card = cc.card
JOIN
    card_holder ch ON cc.cardholder_id = ch.id
JOIN
    merchant m ON t.id_merchant = m.id
WHERE
    ch.id IN (2, 18)
ORDER BY
    t.date DESC;
        """
# Create a DataFrame from the query result. HINT: Use pd.read_sql(query, engine)
df = pd.read_sql(query, engine)
df.tail(10)
```

```python
# Plot for cardholder 2
# Filtering data for cardholder 2
df_cardholder_2 = df[df['cardholder_id'] == 2]

# Plot time series for cardholder 2
plot_cardholder_2 = df_cardholder_2.hvplot.line(x='date', y='amount', title='Cardholder 2 Transactions')

# Display the plot
plot_cardholder_2
```

```python
# Plot for cardholder 18
# Filtering data for cardholder 18
df_cardholder_18 = df[df['cardholder_id'] == 18]

# Plot time series for cardholder 18
plot_cardholder_18 = df_cardholder_18.hvplot.line(x='date', y='amount', title='Cardholder 18 Transactions')

# Display the plot
plot_cardholder_18
```

```python
# Combined plot for card holders 2 and 18
plot_combined = df.hvplot.line(x='date', y='amount', by='cardholder_id', title='Cardholder 2 and 18 Transactions')

# Display the combined plot
plot_combined
```
